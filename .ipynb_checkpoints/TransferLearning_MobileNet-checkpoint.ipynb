Defining the input image size as `IMG_SIZE` and the input shape as `IMG_SHAPE`.

```python
IMG_SIZE  = 224
IMG_SHAPE = (IMG_SIZE, IMG_SIZE, 3)
```

**Defining a helper function** called `prepare_image`: A function that  prepares images for the mobileNet pre-trained model. Input shape: (1, 224, 224, 3)

Available at: https://www.tensorflow.org/api_docs/python/tf/keras/applications/mobilenet/preprocess_input

```python
def prepare_image(img_path, img_size = 224):
    img = image.load_img(img_path, target_size=(img_size, img_size))
    img_array = image.img_to_array(img)
    img_array_expanded_dims = np.expand_dims(img_array, axis=0)
    return preprocess_input(img_array_expanded_dims)
```

**Prepare the data:** Load the data into the ImageDataGenerator. 

Available at: https://www.tensorflow.org/api_docs/python/tf/keras/preprocessing/image/ImageDataGenerator

```python
# Preprocessing the input data
train_datagen=ImageDataGenerator(preprocessing_function=preprocess_input) 

test_datagen=ImageDataGenerator(preprocessing_function=preprocess_input) 


# TRAINING set
listGroupsTrain = os.listdir(train_path) # the directory path


# TEST set
listGroupsTest = os.listdir(test_path) # the directory path

number_groups = len(listGroupsTrain)

##
## Be sure that list of training and testing groups are in the same order
listGroupsTest = listGroupsTrain
##


# Load the data into the ImageDataGenerator
train_generator=train_datagen.flow_from_directory(train_path,
                                                  target_size=(IMG_SIZE,IMG_SIZE),
                                                  color_mode='rgb',
                                                  batch_size=16,
                                                  class_mode='categorical',
                                                  shuffle=True,
                                                  classes=listGroupsTrain)

test_generator=test_datagen.flow_from_directory(test_path,
                                                target_size=(IMG_SIZE,IMG_SIZE),
                                                color_mode='rgb',
                                                batch_size=16,
                                                class_mode='categorical',
                                                shuffle=False,
                                                classes=listGroupsTest)

print('The ', number_groups, 'groups are:',  listGroupsTrain)
print(listGroupsTest)
```




# 3. Transfer Learning

 ### **Transfer the `base` on a new `head`**

This is the transfer learning strategy:
1. Use the pre-trained convolutional layers obtained by mobileNet (`base`)
2. Adjust a new dense layer in the end (`head` or `classifier`)
3. Freeze the base layers and train only the classifier.

In other words, this is transferring the ability of mobileNet to  extract features. The number of  output neurons of the new classifier should be the number of groups (classes) of our new dataset.

Model: https://keras.io/api/applications/

```python
# Loading and inspecting the pre-trained model
mobile_full = MobileNet(weights='imagenet')
mobile_full.summary()
```

Loading the pre-trained mobileNet again and keeping only the `base`: `transferModel_base`

```python
## keeping only the base of the pre-trained model
transferModel_base = MobileNet(input_shape=IMG_SHAPE, weights='imagenet', include_top=False)
```

**Model Building**

Create a new architecture `model` by transfering the `base` of the MobileNet on a new `head`. 

Freeze the convolutional layers and train only the dense layers.

Regularize the dense layer: Regularization prevents the network from overfitting.



```python
# Freeze the mobileNet's convolutional layers
transferModel_base.trainable = False

# Regularize the dense layers 
kernel_weight = 0.005 #9
bias_weight = kernel_weight


model = tf.keras.Sequential([transferModel_base,
                             GlobalAveragePooling2D(),
                             Dense(number_groups, activation='softmax', name='new_head',
                                   kernel_regularizer=regularizers.l1(kernel_weight),
                                   kernel_initializer='random_normal',
                                   bias_regularizer=regularizers.l1(bias_weight),
                                   bias_initializer='uniform')
                            ])
model.summary()
```

**Model Compilation**

Optimizer: https://www.tensorflow.org/api_docs/python/tf/keras/optimizers/Adam

Loss: https://www.tensorflow.org/api_docs/python/tf/keras/metrics/CategoricalCrossentropy

Metrics: https://www.tensorflow.org/api_docs/python/tf/keras/metrics/Accuracy

```python
lrate_head = 0.0025 #2
head_epochs=50

model.compile(optimizer=Adam(lr=lrate_head),
              loss='categorical_crossentropy',
              metrics=['accuracy'])     
```

**Model Training/Fitting**

```python
step_size_train=len(train_generator)

history = model.fit_generator(generator=train_generator,
                              validation_data=test_generator,
                              steps_per_epoch=step_size_train,
                              epochs=head_epochs)
```

**Making a helper function that plots the training/test loss and accuracy**

```python


def plot_metrics(epoch, historyModel):

    epochsVec = np.linspace(1,epoch,epoch)

    acc = historyModel.history['accuracy']
    val_acc = historyModel.history['val_accuracy']

    loss = historyModel.history['loss']
    val_loss = historyModel.history['val_loss']

    plt.figure(figsize=(16, 4))
    plt.subplot(1, 2, 1)
    plt.plot(epochsVec,acc, '-o', label='Training Accuracy')
    plt.plot(epochsVec,val_acc, '-o', label='Validation Accuracy')
    plt.legend(loc='lower right')
    plt.ylabel('Accuracy')
    plt.ylim([min(plt.ylim()),1])
    plt.xlim([1, epoch])
    plt.title('Training and Validation Accuracy')

    plt.subplot(1, 2, 2)
    plt.semilogy(epochsVec,loss,'-o', label='Training Loss')
    plt.semilogy(epochsVec,val_loss,'-o', label='Validation Loss')
    plt.legend(loc='upper right')
    plt.ylabel('Cross Entropy')
    plt.xlim([1, epoch])
    plt.title('Training and Validation Loss')
    plt.xlabel('epoch')
    plt.show()
```

**Trace plot**

```python

plot_metrics(head_epochs, history)
```

**Model Evaluation**

```python
# Get the accuracy score on the Test data
batch_size = 16
test_score = model.evaluate_generator(test_generator, batch_size)
print("The Classification accuracy: {:.2f}%".format(test_score[1] * 100)) 

print(" The Loss: ",test_score[0])
```
